```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\spngu\Downloads\stock_market_performance_analysis_csv.csv")
df.head()
```

```python
# Data cleaning 
df['date']=pd.to_datetime(df['date'])
df.dropna(inplace=True)
df['date'].head()
```

```python
# Python modification (extra analysis)
# Q1:Daily return percentage
df['daily_return_pct']=((df['close']-df['open'])/df['open'])*100
df.head()
```

```python
# Q2:Volatility calculation
df['volatility']=df['high']-df['low']
df.head()
```

```python
# Q3:Trend column
df['trend']=np.where(df['close']>df['open'],'Bullish','Bearish')
df.head()
```

```python
df.to_csv(r"C:\Users\spngu\Downloads\stock_market_performance_analysis_csv.csv",index=False)
```
